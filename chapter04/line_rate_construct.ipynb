```python
import pywt
from scipy.io import loadmat
```

```python
mat = loadmat('data/leleccum.mat')
signal = mat['leleccum'][0]
signal
```

```python
coeffs = pywt.wavedec(signal, 'bior3.7', level=5)
coeffs
```
